### Manual bundle definition

Four lists have to be provided:
* Ranges defining helices
* Chain names
* Helices orientation (True=antiparallel)
* Heptad positions for the first layer (not used, can be any letter)

```python
defdata = [[range(8, 47), range(8, 47), range(8,47)[::-1], range(8,47)[::-1]],
           ['A', 'B', 'C', 'D'],
           [False, False, True, True],
           ['x', 'x', 'x', 'x']]

run_samcc_turbo('./example_in/1qx8.pdb1', './example_out/', mode='defdata', defdata=defdata)
```
