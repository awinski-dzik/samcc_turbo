```python
import os
import warnings
warnings.filterwarnings("ignore")

from samcc.bundleClass import bundleClass
from samcc.wrappers import run_dssp
from samcc.wrappers import run_socket
from samcc.socket_parser import parse_socket_output
from samcc.socketClass import socket_class
```

### Let's analyze <a href="https://www.rcsb.org/structure/3tsi" target="_blank">this</a> structure!


![](./example_in/3tsi_view.png)




```python
pdbpath='./example_in/3tsi.pdb'

dssp_binary = '/opt/apps/dssp-2.3.0/mkdssp' # do not use newer version!
socket_binary = '/opt/apps/socket/socket'
```

```python
# predict secondary structure with DSSP
dssppath = run_dssp(pdbpath, dssp_binary)

# predict coiled-coil regions with SOCKET
socketpath = run_socket(pdbpath, dssppath, socket_binary)

# parse SOCKET data
socket_data = parse_socket_output(socketpath)
s = socket_class(socket_data, pdbpath)

# remove temporary files
os.remove(dssppath)
os.remove(socketpath)
```

```python
# raw SOCKET data suggests that this structure contains two coiled-coil domains
# what is kind of strange, given how the 3tsi structure looks like
len(socket_data)
```

```python
# define a bundle object for each detected domain
bundles = s.get_bundles(res_num_layer_detection=5, distance_threshold='auto', search_layer_setting_num=9) 
```

```python
# bundle object provides various measurement methods 
bundle_dfs = {}

for bundle in bundles: 
    bundle.calc_bundleaxis() # define bundle axis
    bundle.calc_radius() # define bundle radius
    bundle.calc_periodicity() # define helical and bundle periodicity
    bundle_dfs[bundle.socket_cc_id] = bundle.gendf() # export a pandas data frame with all the data
    
print(f'{list(bundle_dfs.keys())} bundles detected...') 
```

```python
# Now, let's take a look at the two domains.
# P and p denote bundle and helical periodicity, respectively.
bundle_dfs['10']
```

```python
bundle_dfs['11']
```

The first domain spans residues 61-72, whereas the second 83-94, but what's is happening between them?

Once we know domain boundaries and layers, we can force SamCC-Turbo to measure also the missing part (i.e, to measure the whole region as it is were a single coiled coil).


```python
bundle = bundleClass()

# we use `defdata` to define the boundaries of this big domain
defdata = [[range(61, 95), range(61, 95), range(61, 95), range(61, 95)], # residue ranges
           ['A', 'B', 'C', 'D'], # PDB chains ordered as in the bundle
           [False, False, False, False], # no anti-parallel helices
           ['x', 'x', 'x', 'x']] # no hepatd register definition

bundle.from_defdata(pdbpath, *defdata)
```

```python
# Now, we do measurements for the whole region:
bundle.calc_bundleaxis() 
bundle.calc_radius() 
bundle.calc_periodicity() 
big_bundle_df = bundle.gendf() 
```

```python
# ... and plot the helical periodicity (averaged per layer):
p_mean = big_bundle_df.p.mean(level=0)
p_mean.plot()
```

The helices of the two coiled-coil segments adopt a canonical periodicity of ~3.63 residues per turn. However, in the linker region, the helical periodicity increases to almost 5 residues per turn, a value characteristic for pi-helices. That's the reason why it was not detected by SOCKET (presumably due to the lack of knobs-into-holes interactions).

```python
print (f'helical periodicity of bundle #1 {p_mean[:17].mean().round(2)} res/turn -> alpha helix')
print (f'helical periodicity of bundle #2 {p_mean[23:].mean().round(2)} res/turn -> alpha helix')
print (f'\nhelical periodicity of the linker {p_mean[17:23].mean().round(2)} res/turn -> pi helix')
```

Take a look at the structure again to localize the pi-helical linker: <a href="https://www.rcsb.org/3d-view/3TSI" target="_blank">3D view at PDB</a>



```python
# Let's investigate the periodicity of the bundle (i.e., supercoiling of the helices):
P_mean = big_bundle_df.P.mean(level=0)
P_mean.plot()
```

```python
print (f'periodicity of bundle #1 {P_mean[:17].mean().round(2)} res/turn -> canonical 7/2 coiled coil')
print (f'periodicity of bundle #2 {P_mean[23:].mean().round(2)} res/turn -> non-canonical 11/3 coiled coil')
```

The 3tsi structure comprises two coiled-coil regions of different periodicities. The first N-terminal domain is a canonical left-handed coiled-coil (7/2=3.5 periodicity), whereas in the second C-terminal, the helices are straight and display only subtle right-handed supercoiling (the difference between bundle periodicity, 3.64, and the periodicity of a regular alpha helix, 3.63, is very small).

Below we will investigate the core packing of the first domain. 

```python
with bundles[0] as b:
    b.calc_crick() # calculates the Circk angle for each residue in the bundle
    b.calc_crickdev(3.5, 7, optimal_ph1=19.5) # calculates the difference between the observed Crick angles 
    # (see above) and expected for an ideal bundle of a given periodicity. The `optimal_ph1` defines
    # the reference rotation state of the helices (see also https://pubmed.ncbi.nlm.nih.gov/20139000/)
    # defined based on many parallel tetrameric structures.
    bundle_dfs[b.socket_cc_id] = b.gendf() # update data frame
```

```python
# This result indicates the core packing of the first domain is fairly canonical (so-called a-d packing)
# (the difference between the observed orientation of the side-chain and one expected for an ideal bundle
# is close to zero)
ax=bundle_dfs['10'].crdev.mean(level=0).plot()
ax.set_ylim(-26, 26)
```
